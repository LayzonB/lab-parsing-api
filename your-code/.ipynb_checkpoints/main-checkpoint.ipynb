## Reading the data

In the next cell, use Python to obtain data from the API endpoint you chose in the previous step. Workflow:

1. Import libraries.

1. Define the initial API endpoint URL.

1. Make request to obtain data of the 1st page. Flatten the data and store it in an empty object variable.

1. Find out the total page count in the 1st page data.

1. Use a FOR loop to make requests for the additional pages from 2 to page count. Append the data of each additional page to the flatterned data object.

1. Print and review the data you obtained.

```python
# your code here
#Import libraries.

import requests
import pandas as pd
import json
import urllib.request
# tuve que usar esta libreria por que no podia parsear el json, me daba error y consegui hacerlo con esta \
# no se si la regue con esata lib. Todo el mundo recomendaba usar la requests en SO pero no supe como componerlo

#Define the initial API endpoint URL.

url="https://www.zalando.fr/api/catalog/articles?age_groups=KIDS&categories=promo-enfant&genders=MALE&limit=84&offset=84"

#make request to obtain data of the 1st page. Flatten the data and store it in an empty object variable.
data = urllib.request.urlopen(url).read().decode()
obj = json.loads(data)
flattened_data = pd.json_normalize(obj)
flattened_data1 = pd.json_normalize(flattened_data.articles[0])
flattened_data1.head(10)
```

```python
obj
```

```python
# Find out the total page count in the 1st page data.

totalpages= obj["pagination"]
totalpages
```

```python
obj["pagination"]['page_count']
```

```python
#no corría con el total de paginas tuve que cortar el loop
pages = 4
df= pd.DataFrame()
for i in range(2,pages):
    k=84*i
    url=f'https://www.zalando.fr/api/catalog/articles?categories=homme&limit=84&offset={k}'
    response = requests.get(url)
    data = urllib.request.urlopen(url).read().decode()
    obj = json.loads(data)    
    flattened_data = pd.json_normalize(obj)
    flattened_data1 = pd.json_normalize(flattened_data.articles[0])
    flattened_data1=flattened_data1.set_index('sku')
    df = df.append(flattened_data1)

df    


```

```python
#esto lo copie del learning pero no me queda claro por que defino la columna con el method de regex

df['price.original']=df['price.original'].str.extract('(\d*,\d*)')
df['price.promotional']=df['price.promotional'].str.extract('(\d*,\d*)')

df['price.promotional']
```

```python
df['price.original'] = [x.replace(',', '.') for x in df['price.original']]
df['price.promotional'] = [x.replace(',', '.') for x in df['price.promotional']]

df['price.promotional']
df['discount_amount']=df['price.original'].astype(float)-df['price.promotional'].astype(float)
```

```python
df1= df.copy()
total_disc=df1.groupby(['brand_name']).sum().discount_amount
total_disc
```
